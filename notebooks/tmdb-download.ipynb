## Connection

```python
from kedro.config import ConfigLoader

conf_paths = ["../conf/local"]
conf_loader = ConfigLoader(conf_paths)
credentials = conf_loader.get("credentials*", "credentials*/**")
```

```python
import tmdbsimple as tmdb
```

```python
tmdb.API_KEY = credentials["tmdb_api"]["key"]
```

## Query

```python
import random
import pandas
```

```python
%%time

n = 10_000
ids = [i for i in range(n)]
movies = []

for i in ids:
    try:
        movie = tmdb.Movies(i).info()
        movies.append(movie)
    except Exception as ex:
        print(ex)
        
```

```python
movies[:1]
```

```python
movie_data = pandas.json_normalize(movies)
```

_let's keep it SFW_

```python
movie_data = movie_data[~movie_data["adult"]]
```

```python
movie_data.to_parquet("movies.parquet")
```
